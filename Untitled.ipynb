```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
import string
import re
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn.feature_extraction.text import TfidfVectorizer
from sklearn.naive_bayes import MultinomialNB
from sklearn.metrics import classification_report,accuracy_score,confusion_matrix
import pickle
from wordcloud import WordCloud
from collections import Counter

from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import accuracy_score, confusion_matrix, classification_report

```

```python
df = pd.read_csv('spam.csv', encoding='latin-1')
```

```python


#  Drop unnecessary columns & rename useful ones
df = df[['v1', 'v2']]  # Keep only the label and text
df.columns = ['label', 'message']

#  Convert labels to binary (ham = 0, spam = 1)
df['label'] = df['label'].map({'ham': 0, 'spam': 1})

#Remove duplicate messages
df = df.drop_duplicates()

# Check & drop nulls if any
df = df.dropna()

#  Clean the text
def clean_text(text):
    text = text.lower()  # Lowercase
    text = re.sub(r'\n', ' ', text)  # Remove newlines
    text = re.sub(r'\W', ' ', text)  # Remove non-word characters
    text = re.sub(r'\s+', ' ', text)  # Remove extra spaces
    text = text.strip()  # Remove leading/trailing spaces
    return text

df['cleaned_message'] = df['message'].apply(clean_text)


print(df.head())
```

```python
df = pd.read_csv('spam.csv', encoding='latin-1')
df = df[['v1', 'v2']]
df.columns = ['label', 'message']
df['label'] = df['label'].map({'ham': 0, 'spam': 1})
df = df.drop_duplicates().dropna()
df['cleaned_message'] = df['message'].str.lower().str.replace(r'\W', ' ', regex=True).str.replace(r'\s+', ' ', regex=True).str.strip()
```

```python
df['message_length'] = df['cleaned_message'].apply(len)
```

```python
def get_common_words(messages, n=20):
    words = ' '.join(messages).split()
    return Counter(words).most_common(n)

common_spam_words = get_common_words(df[df['label'] == 1]['cleaned_message'])
common_ham_words = get_common_words(df[df['label'] == 0]['cleaned_message'])

print("\nTop 10 common words in spam messages:")
for word, count in common_spam_words[:10]:
    print(f"{word}: {count}")

print("\nTop 10 common words in ham messages:")
for word, count in common_ham_words[:10]:
    print(f"{word}: {count}")
```

```python
#Preprocessing
```

```python
X = df['cleaned_message']
y = df['label']
```

```python
tfidf = TfidfVectorizer(max_features=3000)  # You can tune this
X_tfidf = tfidf.fit_transform(X)
```

```python
X_train, X_test, y_train, y_test = train_test_split(X_tfidf, y, test_size=0.2, random_state=42)
```

```python

print("Train Shape:", X_train.shape)
print("Test Shape:", X_test.shape)
```

```python
nb_model = MultinomialNB()
nb_model.fit(X_train, y_train)
nb_pred = nb_model.predict(X_test)
```

```python
lr_model = LogisticRegression()
lr_model.fit(X_train, y_train)
lr_pred = lr_model.predict(X_test)
```

```python

def evaluate_model(name, y_test, y_pred):
    print(f"📊 Model: {name}")
    print("Accuracy:", accuracy_score(y_test, y_pred))
    print("Classification Report:\n", classification_report(y_test, y_pred))
    
    # Confusion Matrix
    cm = confusion_matrix(y_test, y_pred)
    plt.figure(figsize=(4,3))
    sns.heatmap(cm, annot=True, fmt='d', cmap='Blues', xticklabels=['Ham', 'Spam'], yticklabels=['Ham', 'Spam'])
    plt.title(f'Confusion Matrix - {name}')
    plt.xlabel('Predicted')
    plt.ylabel('Actual')
    plt.show()

# ----------- Evaluate All Models -----------
evaluate_model("Naive Bayes", y_test, nb_pred)
evaluate_model("Logistic Regression", y_test, lr_pred)
```
